Question 2: What are the key assumptions of Simple Linear Regression?

The key assumptions of Simple Linear Regression (SLR) are conditions that should be satisfied for the modelâ€™s results to be valid and reliable:

Linearity
The relationship between the independent variable
ğ‘¥
x and the dependent variable
ğ‘¦
y is linear. This means the data points can be reasonably represented by a straight line.

Independence of Errors
The residuals (errors) are independent of each other. This assumption is especially important for time-series data, where observations should not be correlated over time.

Homoscedasticity (Constant Variance)
The variance of the residuals is constant across all values of the independent variable. In other words, the spread of errors remains the same along the regression line.

Normality of Errors
The residuals are normally distributed with a mean of zero. This assumption is mainly important for hypothesis testing and confidence intervals.

No Perfect Multicollinearity
In SLR, this means the independent variable has variation (it is not constant). Without variation in
ğ‘¥
x, a regression line cannot be fitted.

Zero Mean of Errors
The expected value of the error term is zero for all values of
ğ‘¥
x, ensuring unbiased estimates of the regression coefficients.

Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# Independent variable (X)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)

# Dependent variable (y)
y = np.array([2, 4, 6, 8, 10])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```
